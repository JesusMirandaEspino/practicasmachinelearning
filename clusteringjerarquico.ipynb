```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
iris = load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
linked = linkage( X,  'ward')
```

```python
plt.figure(figsize=(10,7))

dendrogram( linked, orientation='top', labels=y, distance_sort='descending', show_leaf_counts=True )

plt.grid(True)
plt.title('Dendograma de Cloustering');
```

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage


iris = load_iris()


X = iris.data
y = iris.target


Z = linkage( X,  'ward')


plt.figure(figsize=(10,7))

dendrogram( Z, orientation='top', labels=y, distance_sort='descending', show_leaf_counts=True )

plt.grid(True)
plt.title('Dendrograma del Agrupamiento Jerrquico - Mtodo de Enlace Ward')
plt.xlabel('ndice de la Muestra')
plt.ylabel('Distancia')
```
