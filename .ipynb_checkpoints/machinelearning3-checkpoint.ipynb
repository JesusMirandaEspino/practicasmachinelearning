```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame()
df['area'] = [26000, 25000, 50000,35000, 40000]
df['precio'] = [126000, 115000,850000,335000, 450000]
```

```python
plt.scatter( df['area'], df['precio']  )
```

```python
X = df[['area']]
```

```python
X
```

```python
y = df['precio']
```

```python
y
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(X,y)
```

```python
modelo.predict([[60000]])
```

```python
plt.scatter( df['area'], df['precio']  )
plt.plot( df['area'], modelo.predict(X) );
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame()
df['Edad'] = [25, 30, 35, 40, 45]
df['Salario'] = [3000, 3500, 4000, 4500, 5000]
```

```python
plt.scatter( df['Edad'], df['Salario'], marker='o' , color='blue'  )
plt.grid(True)
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.title('Relación entre Edad y Salario');

```

```python
X = df[['Edad']]
y = df['Salario']
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(X,y)
```

```python
plt.scatter( df['Edad'], df['Salario'], marker='o' , color='blue'  )
plt.plot( df['Edad'], modelo.predict(X) )
plt.grid(True)
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.title('Relación entre Edad y Salario');
```
