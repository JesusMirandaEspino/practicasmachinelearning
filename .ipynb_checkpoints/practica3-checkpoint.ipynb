```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import seaborn as sns
from sklearn.linear_model import LogisticRegression

from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("C:/Users/jesus/Desktop/curso IA/Ventastienda.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Fecha'] = pd.to_datetime( df['Fecha'])
df['Fecha']
```

```python
df
```

```python
escala = MinMaxScaler(feature_range=(0,1))
```

```python
columnas_para_escalar = df.drop(['Ventas', 'Fecha'], axis=1).columns
```

```python
columnas_para_escalar
```

```python
normado = escala.fit_transform(df[columnas_para_escalar])
```

```python
df_normado = pd.DataFrame( data=normado, columns=columnas_para_escalar )
```

```python
df_normado
```

```python
df_normado['Ventas'] = df['Ventas']
df_normado['Fecha'] = df['Fecha']
```

```python
df_normado
```

```python
X =  df_normado.drop(['Ventas', 'Fecha'], axis=1)
y = df_normado['Ventas']
```

```python
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split( X,  y, train_size=0.8, random_state=42 )
```

```python
df.describe()
```

```python
plt.figure( figsize=(12,6) )
sns.histplot( df['Ventas'], bins=30,kde=True )
plt.title('Dsitribucion de Ventas')
plt.xlabel('Ventas')
plt.ylabel('Frecuencia');
```

```python
plt.figure( figsize=(12,6) )
sns.barplot( data=df, x='DíaDeLaSemana', y='Ventas' )
plt.title('Distribuciond de Ventas por dias de la semana')
plt.xlabel('Dias de la semana')
plt.ylabel('Ventas Promedio');
```

```python
plt.figure( figsize=(12,6) )
sns.boxplot( data=df, x='Promociones', y='Ventas' )
plt.title('Efecto de Promociones en las Ventas')


plt.figure( figsize=(12,6) )
sns.boxplot( data=df, x='Festivo', y='Ventas' )
plt.title('Efecto de dias Festivos en las Ventas')


plt.figure( figsize=(12,6) )
sns.boxplot( data=df, x='Promociones', y='Ventas', hue='Festivo' )
plt.title('Relacion de los dias Festivos y Promociones en las Ventas')

```

```python
modelos= [ ( 'Regresion Logistica', LogisticRegression()),
          ( 'Arbol de Decision', DecisionTreeClassifier( random_state=42)),
          ( 'Bosque Aleatorio', RandomForestClassifier( random_state=42))
         ]
```

```python
for nombre, modelo in modelos:
    modelo.fit(X_entrena, y_entrena)
    puntaje = modelo.score(X_prueba, y_prueba)
    print(f'{nombre}: {puntaje}')
```

```python
modelo_lineal = LinearRegression()
```

```python
modelo_lineal.fit(X_entrena, y_entrena)
```

```python
prediccion_lineal = modelo_lineal.predict(X_prueba)
```

```python
plt.figure( figsize=(10,6) )
plt.scatter( y_prueba, prediccion_lineal, alpha=0.5 )
plt.plot( [y.min(), y.max()], [y.min(), y.max()] )
plt.title('Dsitribucion de Ventas')
plt.xlabel('Ventas Reales')
plt.ylabel('Ventas Predictivas');
```

```python
df_test = pd.DataFrame( { 'Real': y_prueba, 'Predicho': prediccion_lineal } )
df_test = df_test.sort_index()
# X_entrena, X_prueba, y_entrena, y_prueba
```

```python
plt.figure( figsize=(15,5) )
plt.plot(df_test['Real'], label='Ventas Reales' , alpha=0.7 )
plt.plot(df_test['Predicho'], label='Ventas Predicho' , alpha=0.7 )
plt.legend()
plt.title('Comparacion entre Ventas Reales y Ventas Predichas');
```
