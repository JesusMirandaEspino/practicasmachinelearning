```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.utils import resample


datos = {
    'edad': [25, 34, 45, 23, 35],
    'genero': ['F', 'M', 'M', 'F', 'M'],
    'contratado': [1, 0, 1, 0, 1]
}

datos_desbalanceados = pd.DataFrame(datos)


agrupado = datos_desbalanceados.groupby('genero')


data_balanceada  = pd.DataFrame()




for nombre, grupo in agrupado:
    grupo_balandeado = resample(grupo, replace=True, n_samples=10, random_state=123)
    data_balanceada = pd.concat( [data_balanceada, grupo_balandeado] )
```
