```python
import numpy as np
import random
```

```python
dimensions = (5,5)
estado_inicial = (0,0)
meta = (4,4)
obstaculos = [ (1,1), (1,3), (2,3), (3,0) ]
acciones = [(-1,0), (1,0), (0,-1), (0,1)]
```

```python
num_estados = dimensions[0] * dimensions[1]
num_estados
```

```python
numeros_acciones = len(acciones)
numeros_acciones
```

```python
Q = np.zeros((num_estados, numeros_acciones))
Q
```

```python
dimensions
```

```python
def estadoIndice(estado):
    return estado[0] * dimensions[1] + estado[1]
```

```python
ejemplo = estadoIndice((1,0))
ejemplo
```

```python
alpha = 0.1
gama = 0.99
epsilon = 0.2
episodios = 100
```

```python
def elegir_accion(estado):
    if random.uniform(0,1) < epsilon:
        return random.choice( range( numeros_acciones ))
    else: 
     return np.argmax(Q[estadoIndice(estado)] )
```

```python
def aplicar_accion(estado, accion_idx):
    accion = acciones[accion_idx]
    nuevo_estado = tuple(np.add( estado, accion ) % dimensions )

    if nuevo_estado in obstaculos or nuevo_estado == estado:
        return estado, -100, False
    if nuevo_estado == meta:
        return nuevo_estado, 100, True
    return nuevo_estado, -1, False

    
```

```python
for episodio in range(episodios):
    estado = estado_inicial
    terminado = False

    while not terminado:
        idx_estado = estadoIndice(estado)
        print(f"el estado { idx_estado}")
        accion_idx = elegir_accion(estado)
        print(f"la accion { accion_idx}")
        nuevo_estado, recompensa, terminado = aplicar_accion(estado, accion_idx)
        print(f"Nuevo estado { nuevo_estado}, Recompensa, {recompensa}, terminado {terminado}")
        idx_nuevo_estado = estadoIndice(nuevo_estado)
        Q[ idx_estado, accion_idx] = Q[ idx_estado, accion_idx] + alpha * ( recompensa + gama * np.max( Q[idx_nuevo_estado, ] ) - Q[ idx_estado, accion_idx]  )
        print(f"el valor de Q es { Q[ idx_estado, accion_idx]}")
        estado = nuevo_estado
```

```python
politica = np.zeros( dimensions, dtype=int )
politica
```

```python
for i in range(dimensions[0]):
    for j in range(0,dimensions[0] ):
        estado = (i,j)
        idx_estado = estadoIndice(estado)
        mejor_accion = np.argmax(Q[idx_estado])
        politica[i,j] = mejor_accion
print( "Plitica aprendida (0) arriba, (1) abajo, (2) izquierda  y (3)derecha" )
politica
```
