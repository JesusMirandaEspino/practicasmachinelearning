```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
modelo = RandomForestClassifier( n_estimators=100, random_state=42 )
```

```python
puntaje = cross_val_score(modelo, X, y, cv=5)
```

```python
puntaje
```

```python
puntaje.mean()
```

```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
modelo = RandomForestClassifier(random_state=42 )
```

```python
parametros = { 'n_estimators': [50,100,200],
                'max_features': [ 'sqrt', 'log2' ],
                 'max_depth': [ 4, 5, 6, 7, 8 ],
              'criterion': ['gini', 'entropy']
             }
```

```python
mi_grid_search = GridSearchCV( estimator=modelo, param_grid=parametros, cv=5, scoring='accuracy' )
```

```python
mi_grid_search.fit(X, y)
```

```python
mi_grid_search.best_params_
```

```python
mi_grid_search.best_score_
```
