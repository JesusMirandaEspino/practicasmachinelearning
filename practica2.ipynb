```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import seaborn as sns
```

```python
data = {
    'feature1': [1.2, 2.5, 3.8, 4.1, 5.6, 6.0, 7.3, 8.9, 9.2, 10.5],
    'feature2': [2.1, 3.4, 4.7, 5.0, 6.5, 7.9, 8.2, 9.5, 10.8, 11.1],
    'feature3': [0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0],
    'label': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A', 'B']
}
```

```python
df = pd.DataFrame(data)
```

```python
X = df.drop('label', axis=1)
```

```python
le = LabelEncoder()
```

```python
tipo = le.fit_transform( df['label'] )
tipo
```

```python
y = tipo
```

```python
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split( X,  y, train_size=0.8, random_state=42 )
```

```python
arbol = DecisionTreeClassifier()
```

```python
arbol.fit(X_entrena, y_entrena)
```

```python
plot_tree(decision_tree=arbol);
```

```python
plt.figure(figsize=(20, 20))
plot_tree(decision_tree=arbol,
        filled = True,
          feature_names=df.columns.to_list(),
        class_names=["A", "B"]);
```
