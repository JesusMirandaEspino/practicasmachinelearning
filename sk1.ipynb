```python
from sklearn import datasets
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
X_entrena, X_prueba, y_entrena, y_prueba =  train_test_split( X, y, test_size=0.25, random_state=0 )
```

```python
scaler_ = StandardScaler()
```

```python
scaler_.fit(X_entrena)
```

```python
X_entrena_scalado = scaler_.transform( X_entrena )
X_prueba_scalado = scaler_.transform( X_prueba )
```

```python
modelo = LogisticRegression()
```

```python
modelo.fit(X_entrena_scalado, y_entrena)
```

```python
y_pred = modelo.predict(X_prueba_scalado)
puntaje = modelo.score( X_prueba_scalado,  y_prueba) 
```
