```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
np.random.seed(42)
```

```python
data = np.random.rand(100, 2)
```

```python
data
```

```python
kmeans = KMeans( n_clusters=5, random_state=42 )
kmeans.fit(data)
```

```python
centroides = kmeans.cluster_centers_
centroides
```

```python
etiquetas = kmeans.labels_
etiquetas
```

```python
plt.figure(figsize=(10,6))

plt.scatter( data[:,0], data[:,1], c=etiquetas, cmap='viridis', marker='o'  )
plt.scatter( centroides[:,0], centroides[:,1], c='red', s=200, marker='x', label='Centroides'  )
plt.grid(True)
plt.legend()
plt.title('Ejemplo de K-means Clusterings');
```

```python
pinguinos = sns.load_dataset("penguins")
pinguinos.dropna(inplace=True)
pinguinos
```

```python
data =  pinguinos[ [ 'bill_length_mm',  'bill_depth_mm'] ]
data
```

```python
kmeans = KMeans( n_clusters=3, random_state=42 )
kmeans.fit(data)
```

```python
centroides = kmeans.cluster_centers_
centroides
```

```python

etiquetas = kmeans.labels_
etiquetas
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(data=pinguinos, x='bill_length_mm', y='bill_depth_mm', hue=etiquetas, palette='viridis')
plt.scatter( centroides[:,0], centroides[:,1], c='red', s=200, marker='x', label='Centroides'  )
plt.grid(True)
plt.legend()
plt.title('Ejemplo de K-means Clusterings');
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans


np.random.seed(42)

datos = np.random.rand(100, 2)
k=3

kmeans = KMeans( n_clusters=k, random_state=42 )
kmeans.fit(datos)


centroides = kmeans.cluster_centers_
etiquetas = kmeans.labels_

plt.figure(figsize=(10,6))

plt.scatter( datos[:,0], datos[:,1], c=etiquetas, cmap='viridis', marker='o'  )
plt.scatter( centroides[:,0], centroides[:,1], c='red', s=200, marker='x', label='Centroides'  )
plt.grid(True)
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('K-Means Clustering')
```
